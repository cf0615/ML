```python
#import the modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ydata_profiling import ProfileReport
import numpy as np
```

```python
#load data
dataset = pd.read_csv('./Breast_Cancer.csv')

dataset.info()
dataset.head()
```

```python
# Check for missing values in each column
missing_values = dataset.isnull().sum()

# Visualize missing values using a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(dataset.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values in Dataset')
plt.show()

# Print missing values count
print(missing_values)
```

```python
# Get a statistical summary of numerical columns
dataset.describe()
```

```python
# Do a fast eda using ydata_profiling
Report=ProfileReport(dataset,title="Breast Cancer Dataset")
Report.to_notebook_iframe()
```

```python
print(f"Number of rows before dropping duplicates: {dataset.shape[0]}")

# Drop duplicate rows
dataset = dataset.drop_duplicates()

# Confirm the change
print(f"Number of rows after dropping duplicates: {dataset.shape[0]}")
```

```python
#print unique value
for i in dataset.columns:
    uniValues = np.unique(dataset[i])
    print(i)
    print(uniValues)
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# List of columns to be label encoded
label_columns = ["Race", "Marital Status", "T Stage ", "N Stage", "6th Stage", "differentiate", "Grade"]

# Apply LabelEncoder to each of the label columns
for col in label_columns:
    dataset[col] = label_encoder.fit_transform(dataset[col])
    
dataset.head()
```

```python
#one hot encoding
# List of columns to be one-hot encoded
one_hot_columns = ["A Stage", "Estrogen Status", "Progesterone Status"]

# Apply One-Hot Encoding to the selected columns
dataset = pd.get_dummies(dataset, columns=one_hot_columns, drop_first=True)

dataset.head()
```

```python
# Convert boolean columns (True/False) to integers (1/0)
dataset = dataset.applymap(lambda x: 1 if x is True else (0 if x is False else x))

dataset.head()
```

```python
dataset.isnull().sum()
```

```python
# Create box plots to visualize relationships between the first few features and the target
for i in dataset[dataset.columns[:5]]:
    sns.catplot(x=i, y='Status', data=dataset, height=5, aspect=2, kind='box')
    plt.title(f'Relationship between {i} and Status')
    plt.show()
```

```python
# Convert 'Status' column to numeric values ('Alive' -> 0, 'Dead' -> 1)
dataset['Status'] = dataset['Status'].apply(lambda x: 0 if x == 'Alive' else 1)

# Calculate correlation matrix
corr_matrix = dataset.corr()

# Plot the heatmap
f, ax = plt.subplots(figsize=(16, 12))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap with Status Encoded (Alive = 0, Dead = 1)")
plt.show()
```

```python
#save the dataset
dataset.to_csv('preprocessed_dataset.csv', index=False)
```
